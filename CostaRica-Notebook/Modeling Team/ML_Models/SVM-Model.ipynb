```python
import pandas as pd

# Path to your CSV file
csv_file_path = 'NDMN-Notebooks/CostaRica-Notebook/Costa_Rica_Data/Costa_Classification_Data_Cleaned2.csv'

# Load the CSV into a DataFrame
df = pd.read_csv(csv_file_path)

# Display the first few rows
df.head()
```

```python
df = df.drop(columns=['plotid', 'sampleid', 'geometry', 'Another Class', 'Win_Lose', 'Changes15_', 'Water', 'WetLand', 'Cultivation', 'Grass', 'Herbaceous','SAF'])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')

categorical_features = ['Vegetation','Cover', 'Terrain']

ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
encoded_features = ohe.fit_transform(df[categorical_features])
df = pd.concat([df.drop(columns=categorical_features), encoded_features], axis=1)
```

```python
df.head()
```

```python
y = df['Use']  # Set the target variable
X = df.drop(columns=['Use'])  # Drop the target column to get the features

X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=30, test_size=0.2)
```

```python
# List columns with missing values and the count of missing values
missing_values = X.isnull().sum()

missing_values[missing_values > 0]
```

Linear Regression and SVM do not accept missing values encoded as NaN

```python
from sklearn.impute import SimpleImputer

# Columns to impute with the mean (you can customize this if needed)
columns_to_impute = ['BLUE', 'GREEN', 'NIR', 'RED', 'SWIR1', 'SWIR2', 
                     'brightness', 'diff', 'savi', 'topDiv', 'wetness',
                     'mTPI', 'ndvi']

# Initialize the imputer to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Apply the imputer to the selected columns
X[columns_to_impute] = imputer.fit_transform(X[columns_to_impute])

# Drop columns with too many missing values (like altura2, evi, hand30_100)
X = X.drop(columns=['altura2', 'evi', 'hand30_100'])

# Make sure y corresponds to the new X after any row drop
y = y[X.index]  # In case any rows were dropped from X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=30, test_size=0.2)
```

```python
# List columns with missing values and the count of missing values
missing_values = X.isnull().sum()

missing_values[missing_values > 0]
```

```python
from sklearn.svm import SVC

model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
